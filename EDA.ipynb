# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('balanced_dataset.csv')
df.head()
```

```python
# transform data for visualization
vis = df.copy()

vis.loc[vis['Diabetes_binary'] == 0, 'Diabetes_binary'] = 'No Diabetes'
vis.loc[vis['Diabetes_binary'] == 1, 'Diabetes_binary'] = 'Diabetes'

vis.loc[vis['HighBP'] == 0, 'HighBP'] = 'No HighBP'
vis.loc[vis['HighBP'] == 1, 'HighBP'] = 'High BP'

vis.loc[vis['HighChol'] == 0, 'HighChol'] = 'No High Cholesterol'
vis.loc[vis['HighChol'] == 1, 'HighChol'] = 'High Cholesterol'

vis.loc[vis['Smoker'] == 0, 'Smoker'] = 'No'
vis.loc[vis['Smoker'] == 1, 'Smoker'] = 'Yes'

vis.loc[vis['Stroke'] == 0, 'Stroke'] = 'No'
vis.loc[vis['Stroke'] == 1, 'Stroke'] = 'Yes'

vis.loc[vis['HeartDiseaseorAttack'] == 0, 'HeartDiseaseorAttack'] = 'No'
vis.loc[vis['HeartDiseaseorAttack'] == 1, 'HeartDiseaseorAttack'] = 'Yes'

vis.loc[vis['PhysActivity'] == 0, 'PhysActivity'] = 'No'
vis.loc[vis['PhysActivity'] == 1, 'PhysActivity'] = 'Yes'

vis.loc[vis['Fruits'] == 0, 'Fruits'] = 'No'
vis.loc[vis['Fruits'] == 1, 'Fruits'] = 'Yes'

vis.loc[vis['Veggies'] == 0, 'Veggies'] = 'No'
vis.loc[vis['Veggies'] == 1, 'Veggies'] = 'Yes'

vis.loc[vis['HvyAlcoholConsump'] == 0, 'HvyAlcoholConsump'] = 'No'
vis.loc[vis['HvyAlcoholConsump'] == 1, 'HvyAlcoholConsump'] = 'Yes'

vis.loc[vis['AnyHealthcare'] == 0, 'AnyHealthcare'] = 'No'
vis.loc[vis['AnyHealthcare'] == 1, 'AnyHealthcare'] = 'Yes'

vis.loc[vis['NoDocbcCost'] == 0, 'NoDocbcCost'] = 'No'
vis.loc[vis['NoDocbcCost'] == 1, 'NoDocbcCost'] = 'Yes'

vis.loc[vis['GenHlth'] == 1, 'GenHlth'] = 'Excellent'
vis.loc[vis['GenHlth'] == 2, 'GenHlth'] = 'Very Good'
vis.loc[vis['GenHlth'] == 3, 'GenHlth'] = 'Good'
vis.loc[vis['GenHlth'] == 4, 'GenHlth'] = 'Fair'
vis.loc[vis['GenHlth'] == 5, 'GenHlth'] = 'Poor'
vis['GenHlth'] = pd.Categorical(vis['GenHlth'], categories=['Excellent', 'Very Good', 'Good', 'Fair', 'Poor'], ordered=True)

vis.loc[vis['DiffWalk'] == 0, 'DiffWalk'] = 'No'
vis.loc[vis['DiffWalk'] == 1, 'DiffWalk'] = 'Yes'

vis.loc[vis['Sex'] == 0, 'Sex'] = 'Female'
vis.loc[vis['Sex'] == 1, 'Sex'] = 'Male'

vis.loc[vis['Education'] == 1, 'Education'] = 'Never Attended School'
vis.loc[vis['Education'] == 2, 'Education'] = 'Elementary'
vis.loc[vis['Education'] == 3, 'Education'] = 'Junior High School'
vis.loc[vis['Education'] == 4, 'Education'] = 'Senior High School'
vis.loc[vis['Education'] == 5, 'Education'] = 'Undergraduate Degree'
vis.loc[vis['Education'] == 6, 'Education'] = 'Magister'
vis['Education'] = pd.Categorical(vis['Education'], categories=['Never Attended School', 'Elementary', 'Junior High School', 'Senior High School', 'Undergraduate Degree', 'Magister'], ordered=True)

vis.loc[vis['Income'].isin([1, 2, 3, 4]), 'Income'] = 'Less Than $10,000'
vis.loc[vis['Income'].isin([5, 6, 7]), 'Income'] = 'Less Than $35,000'
vis.loc[vis['Income'] == 8, 'Income'] = '$75,000 or More'
vis['Income'] = pd.Categorical(vis['Income'], categories=['Less Than $10,000', 'Less Than $35,000', '$75,000 or More'], ordered=True)

vis.head()
```

```python
vis.info()
```

```python
# Helper functions
def count_and_percentage(col):
    counts = vis[col].value_counts()
    percentages = vis[col].value_counts(normalize=True) * 100
    
    data = pd.concat([counts, percentages], axis=1)
    data.columns = ['Count', 'Percentage']
    print(data)

def catplot_percentage(col):
    sns.countplot(x=col, hue='Diabetes_binary', data=vis)
    plt.xlabel(col)
    plt.ylabel('Count')
    
    # add percentage labels
    total = float(len(df))
    for p in plt.gca().patches:
        height = p.get_height()
        plt.gca().annotate('{:.1f}%'.format((height/total)*100), 
                            xy=(p.get_x()+p.get_width()/2, height),
                            ha='center', va='bottom')

    plt.show()

def numplot(col):
    fig, ax = plt.subplots(3, 1, figsize=(12,10))
    sns.histplot(x=col, hue='Diabetes_binary', data=vis, ax=ax[0])
    sns.boxplot(x=col, y='Diabetes_binary', data=vis, ax=ax[1])
    sns.violinplot(x=col, y='Diabetes_binary', data=vis, ax=ax[2])
    plt.show()
```
